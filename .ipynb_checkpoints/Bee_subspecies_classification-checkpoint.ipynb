Setting the parameter values which are commom

```python
IMAGE_PATH = 'bee_imgs /bee_imgs/'
IMAGE_WIDTH = 100
IMAGE_HEIGHT = 100
IMAGE_CHANNELS = 3
RANDOM_STATE = 2018
TEST_SIZE = 0.20 
VAL_SIZE = 0.20
PATIENCE = 5
VERBOSE = 1
```

##### Loading the data labels

```python
honey_bee_df=pd.read_csv('bee_data.csv')
print(honey_bee_df.shape)
```

##### Loading the images

```python
image_files = list(os.listdir(IMAGE_PATH))
print("Number of image files: {}".format(len(image_files)))
```

```python
file_names = list(honey_bee_df['file'])
print("Matching image names: {}".format(len(set(file_names).intersection(image_files))))
```

Reading the image in the form of an array

```python
def read_image_sizes(file_name):
    image = skimage.io.imread(IMAGE_PATH + file_name)
    return list(image.shape)
```

```python
m = np.stack(honey_bee_df['file'].apply(read_image_sizes))
df = pd.DataFrame(m,columns=['w','h','c'])
honey_bee_df = pd.concat([honey_bee_df,df],axis=1, sort=False)
```

#### For this task , subspecies class is being used. 

```python
categories = (honey_bee_df.groupby(["subspecies"])["subspecies"].nunique()).index
print(categories) # Labels
```

Visualization of all categories(labels) of the class "subspecies"

```python
def draw_category_images(var,cols=5):
    categories = (honey_bee_df.groupby([var])[var].nunique()).index
    f, ax = plt.subplots(nrows=len(categories),ncols=cols, figsize=(2*cols,2*len(categories)))
    # draw a number of images for each location
    for i, cat in enumerate(categories):
        sample = honey_bee_df[honey_bee_df[var]==cat].sample(cols)
        for j in range(0,cols):
            file=IMAGE_PATH + sample.iloc[j]['file']
            im=imageio.imread(file)
            ax[i, j].imshow(im, resample=True)
            ax[i, j].set_title(cat, fontsize=9)  
    plt.tight_layout()
    plt.show()
```

```python
draw_category_images("subspecies")
```

#### Splitting into train and test

```python
train_df, test_df = train_test_split(honey_bee_df, test_size=TEST_SIZE, random_state=RANDOM_STATE, 
                                     stratify=honey_bee_df['subspecies'])
```

#### Splitting train data into train and validation sets

```python
train_df, val_df = train_test_split(train_df, test_size=VAL_SIZE, random_state=RANDOM_STATE, stratify=train_df['subspecies'])
```

```python
print("Train set rows: {}".format(train_df.shape[0]))
print("Test  set rows: {}".format(test_df.shape[0]))
print("Val   set rows: {}".format(val_df.shape[0]))
```

```python
def read_image(file_name):
    image = skimage.io.imread(IMAGE_PATH + file_name)
    image = skimage.transform.resize(image, (IMAGE_WIDTH, IMAGE_HEIGHT), mode='reflect')
    return image[:,:,:IMAGE_CHANNELS]
```

##### Arranging the data in terms of subspecies

```python
def categories_encoder(dataset, var='subspecies'):
    X = np.stack(dataset['file'].apply(read_image))
    y = pd.get_dummies(dataset[var], drop_first=False)
    return X, y
```

```python
X_train, y_train = categories_encoder(train_df)
X_val, y_val = categories_encoder(val_df)
X_test, y_test = categories_encoder(test_df)
```

### Build the model

```python
model1=Sequential()
model1.add(Conv2D(16, kernel_size=3, input_shape=(IMAGE_WIDTH, IMAGE_HEIGHT,IMAGE_CHANNELS), activation='relu', padding='same'))
model1.add(MaxPool2D(2))
model1.add(Conv2D(16, kernel_size=3, activation='relu', padding='same'))
model1.add(Flatten())
model1.add(Dense(y_train.columns.size, activation='softmax'))
model1.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model1.summary()
```

```python
image_generator = ImageDataGenerator(
        featurewise_center=False,
        samplewise_center=False,
        featurewise_std_normalization=False,
        samplewise_std_normalization=False,
        zca_whitening=False,
        rotation_range=180,
        zoom_range = 0.1, 
        width_shift_range=0.1,
        height_shift_range=0.1, 
        horizontal_flip=True,
        vertical_flip=True)
image_generator.fit(X_train)
```

Loading tensorboard

```python
%load_ext tensorboard
%tensorboard --logdir {logs_base_dir}
logs_base_dir = "./logs"
os.makedirs(logs_base_dir, exist_ok=True)
```

Callback functions for data analysis and saving the model weights

```python
tensorboard_callback_model1 = tf.keras.callbacks.TensorBoard(
    log_dir='logs', histogram_freq=0, write_graph=True,
    write_images=False, update_freq='epoch', profile_batch=2,
    embeddings_freq=0, embeddings_metadata=None)
checkpoint_dir = './checkpoints'
# Name of the checkpoint files
checkpoint_prefix = os.path.join(checkpoint_dir, 'checkpoint_model1.h5')
checkpoint_model1=tf.keras.callbacks.ModelCheckpoint(
    filepath=checkpoint_prefix,
    save_weights_only=True)
```

```python
BATCH_SIZE = 16
NO_EPOCHS = 5
train_model1  = model1.fit_generator(image_generator.flow(X_train, y_train, batch_size=BATCH_SIZE),
                        epochs=NO_EPOCHS,
                        validation_data=(X_val, y_val),
                        steps_per_epoch=len(X_train)/BATCH_SIZE, callbacks=[tensorboard_callback_model1, checkpoint_model1])
logdir = os.path.join(logs_base_dir, datetime.datetime.now().strftime("%Y%m%d-%H%M%S"))
```

The accuracy and loss vs epoch plot is taken from tensorboard and also from the history saved in the inbult function of tensorflow: model.history()

#### Evaluate the model using the test set

```python
def test_accuracy_report(model):
    predicted = model.predict(X_test)
    test_predicted = np.argmax(predicted, axis=1)
    test_truth = np.argmax(y_test.values, axis=1)
    print(metrics.classification_report(test_truth, test_predicted, target_names=y_test.columns)) 
    test_res = model.evaluate(X_test, y_test.values, verbose=0)
    print('Loss function: %s, accuracy:' % test_res[0], test_res[1])
```

```python
test_accuracy_report(model1)
```

Weighted F1 score : 0.82

### Model2
- Adding two Dropout layers, with a probabilty to keep the layer to 0.4. This is done to prevent the model from overfitting, since an additional CNN layer is added.
- Changing filters of the second convolution layer and adding another CNN layer
- Using RMSprop optimizer with a learning rate of 0.001

```python
## Using RMSprop optimizer
opt1 = tf.keras.optimizers.RMSprop(
    learning_rate=0.001,
    rho=0.9,
    momentum=0.0,
    epsilon=1e-07,
    centered=False,
    name="RMSprop")
```

```python
model2=Sequential()
model2.add(Conv2D(16, kernel_size=3, input_shape=(IMAGE_WIDTH, IMAGE_HEIGHT,IMAGE_CHANNELS), activation='relu', padding='same'))
model2.add(MaxPool2D(2))
model2.add(Conv2D(32, kernel_size=3, activation='relu', padding='same'))
model2.add(Dropout(0.4))
model2.add(Conv2D(64, kernel_size=3, activation='relu', padding='same'))
model2.add(MaxPool2D(2))
model2.add(Dropout(0.4))
model2.add(Conv2D(128, kernel_size=3, activation='relu', padding='same'))
model2.add(Flatten())
model2.add(Dense(y_train.columns.size, activation='softmax'))
model2.compile(opt1 , loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
tensorboard_callback_model2 = tf.keras.callbacks.TensorBoard(
    log_dir='logs2', histogram_freq=0, write_graph=True,
    write_images=False, update_freq='epoch', profile_batch=2,
    embeddings_freq=0, embeddings_metadata=None)
checkpoint_dir = './checkpoints'
# Name of the checkpoint files
checkpoint_prefix = os.path.join(checkpoint_dir, 'checkpoint_model2.h5')
checkpoint_model2=tf.keras.callbacks.ModelCheckpoint(
    filepath=checkpoint_prefix,
    save_weights_only=True)
```

```python
BATCH_SIZE = 64
NO_EPOCHS = 100
train_model2  = model2.fit_generator(image_generator.flow(X_train, y_train, batch_size=BATCH_SIZE),
                        epochs=NO_EPOCHS,
                        validation_data=(X_val, y_val),
                        steps_per_epoch=len(X_train)/BATCH_SIZE, callbacks=[tensorboard_callback_model2, checkpoint_model2])
```

### Model4 : 
- Using **early stopping** method: It is a method to reduce computation time, the parameter checks whether there is significant change in the training and validation loss parameters given the **patience** parameter.  
- Since, the model2 has given good results compared to others, this fine tuning is performed on it.
- Early stopping and patience parameter values were referred from [https://www.kaggle.com/gpreda/honey-bee-subspecies-classification]

```python
opt1 = tf.keras.optimizers.RMSprop(
    learning_rate=0.001,
    rho=0.9,
    momentum=0.0,
    epsilon=1e-07,
    centered=False,
    name="RMSprop")
```

```python
model4=Sequential()
model4.add(Conv2D(16, kernel_size=3, input_shape=(IMAGE_WIDTH, IMAGE_HEIGHT,IMAGE_CHANNELS), activation='relu', padding='same'))
model4.add(MaxPool2D(2))
model4.add(Conv2D(32, kernel_size=3, activation='relu', padding='same'))
model4.add(Dropout(0.4))
model4.add(Conv2D(64, kernel_size=3, activation='relu', padding='same'))
model4.add(MaxPool2D(2))
model4.add(Dropout(0.4))
model4.add(Conv2D(128, kernel_size=3, activation='relu', padding='same'))
model4.add(Flatten())
model4.add(Dense(y_train.columns.size, activation='softmax'))
model4.compile(opt1 , loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
earlystopper = EarlyStopping(monitor='loss', patience=5, verbose=1)
tensorboard_callback_model4 = tf.keras.callbacks.TensorBoard(
    log_dir='logs4', histogram_freq=0, write_graph=True,
    write_images=False, update_freq='epoch', profile_batch=2,
    embeddings_freq=0, embeddings_metadata=None)
checkpoint_dir = './checkpoints'
# Name of the checkpoint files
checkpoint_prefix = os.path.join(checkpoint_dir, 'checkpoint_model4.h5')
checkpoint_model4=tf.keras.callbacks.ModelCheckpoint(
    filepath=checkpoint_prefix,
    save_weights_only=True)
```

```python
BATCH_SIZE = 64
NO_EPOCHS = 100
train_model4  = model4.fit_generator(image_generator.flow(X_train, y_train, batch_size=BATCH_SIZE),
                        epochs=NO_EPOCHS,
                        validation_data=(X_val, y_val),
                        steps_per_epoch=len(X_train)/BATCH_SIZE , callbacks=[tensorboard_callback_model4 , earlystopper , checkpoint_model4 ])

logdir = os.path.join(logs_base_dir, datetime.datetime.now().strftime("%Y%m%d-%H%M%S"))
```

```python
print(train_model4.history.keys())
```

## Comparing the model performance with pre-trained model, using Transfer Learning

Here vgg16 model , which is pre- trained on ImageNET dataset, is used for transfer learning. To not tamper with the weights , all the layers are frozen. The last layer is added as a sequential addition to the frozen model  to a max pool layer for flattening the features and a Fully Connected Layer having 7 neurons, as per the number of Layers, with the softmax activation function. The number of epochs are set to 100, to give equal advantage as model4, but early stopping is not implemented.

Loading the training data

```python
X_train, y_train = categories_encoder(train_df)
```

```python
X_train.shape
```

```python
image_generator = ImageDataGenerator(
        featurewise_center=False,
        samplewise_center=False,
        featurewise_std_normalization=False,
        samplewise_std_normalization=False,
        zca_whitening=False,
        rotation_range=180,
        zoom_range = 0.1, 
        width_shift_range=0.1,
        height_shift_range=0.1, 
        horizontal_flip=True,
        vertical_flip=True)
image_generator.fit(X_train)
```

```python
IMG_SHAPE = (100, 100, 3)
feature_extractor = tf.keras.applications.VGG16(input_shape=IMG_SHAPE,
                                                include_top=False,
                                                weights='imagenet')
```

```python
len(feature_extractor.layers)
```

```python
feature_extractor.trainable = False

# Loading the base model architecture 
feature_extractor.summary()
```

Non-trainable params = 0

```python
model5 = tf.keras.Sequential([
  feature_extractor,keras.layers.GlobalAveragePooling2D(),
  keras.layers.Dense(7, activation='softmax')
])

model5.compile(optimizer=tf.keras.optimizers.RMSprop(lr=0.01),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model5.summary()
```

```python
tensorboard_callback_model5 = tf.keras.callbacks.TensorBoard(
    log_dir='logs5', histogram_freq=0, write_graph=True,
    write_images=False, update_freq='epoch', profile_batch=2,
    embeddings_freq=0, embeddings_metadata=None)
checkpoint_dir = './checkpoints'
# Name of the checkpoint files
checkpoint_prefix = os.path.join(checkpoint_dir, 'checkpoint_model5.h5')
checkpoint_model5=tf.keras.callbacks.ModelCheckpoint(
    filepath=checkpoint_prefix,
    save_weights_only=True)
```

```python
BATCH_SIZE = 64
NO_EPOCHS = 100
train_model5 = model5.fit_generator(image_generator.flow(X_train, y_train, batch_size=BATCH_SIZE),
                              steps_per_epoch = len(X_train)/BATCH_SIZE,
                              epochs=100,
                              validation_data=(X_val, y_val),
                              validation_steps=len(X_val)/BATCH_SIZE,
                              callbacks=[tensorboard_callback_model5, checkpoint_model5])
logdir = os.path.join(logs_base_dir, datetime.datetime.now().strftime("%Y%m%d-%H%M%S"))                              
```

```python
f , ax = plt.subplots(1, 2,figsize=(12,5))
ax[0].plot(train_model5.history["loss"], 'r' , train_model5.history["val_loss"] ,'g-')
ax[0].set_title("loss vs epochs")
ax[0].set_xlabel("No of Epochs")
ax[0].set_ylabel("loss")

ax[0].grid()
ax[1].plot(train_model5.history["accuracy"], 'r' , train_model5.history["val_accuracy"] ,'g-')
ax[1].set_title("accuracy vs epochs")
ax[1].set_xlabel("No of Epochs")
ax[1].set_ylabel("accuracy")

ax[1].grid()
```
